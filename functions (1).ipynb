```python
# Lambda fuction
#Lambda function is known as Ananymous Function
#Ananymous Function Means -> A function without a function name
#syntax of lambda function : lambda arguments : expression

```

```python
 f = lambda a,b : a + b

print(f(10,20))
```

```python
n = 2

f = lambda x,y : x*y

for i in range(1,11):
  print(f'{n} * {i} = {f(n,i)}')
```

```python
# map syntax = map(function_name,arguments)

def fun(b):
  if b%2 == 0:
    return b



sol = map(fun,[10,11,12,13,14,15])
print(list(sol))
```

```python
#filter syntax = filter(function_name,arguments)
def fun(b):
  if b%2 == 0:
    return b


sol = filter(fun,[10,11,12,13,14,15])
print(list(sol))
```

```python
#reduce syntax = functools.reduce(function_name,arguments)
import functools
f = [1,4,7,9,3,6]

a = functools.reduce(lambda x,y :x + y , [1,4,7,9,3,6])
print(a)
```

```python
#map with filter
a = map(lambda x:x+20 , filter(lambda x:x%2==0,[10,11,21,22,26,36,35]))
print(list(a))
```

```python
#filter with map
a = filter(lambda x:x%2==0 , map(lambda x:x+20,[10,11,21,22,26,36,35]))
print(list(a))
```

```python
#reduce with map and filter
a = functools.reduce(lambda x,y:x+y,map(lambda x:x+20 , filter(lambda x:x%2==0,[10,11,21,22,26,36,35])))
print(a)
```
