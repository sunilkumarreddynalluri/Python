```python
# class = it is the place where varibales and functions are stored
# object = incharge of a class

class CSE:
  john = 20
  peter = 30

anil = CSE() # anil is an object
print(anil.john)
```

```python
# functions in class

class CSE:
  f = 100

  def fun():
    print('GM')


anil = CSE()
print(anil.f)
```

```python
# functions in class

class CSE:
  f = 100

  def sharuk(b):   # b is a varibale which is holding object (anil) copy
    print(b)
    print('GM')


anil = CSE()
print(anil)
print(anil.f)
anil.sharuk()
```

```python
class CSE:
  f = 100

  def sharuk(self):   # self is a varibale which is holding object (anil) copy
    print('GM')


anil = CSE()
print(anil)
print(anil.f)
anil.sharuk()
```

```python
class MECH:
  r = 200

  def kumar(self): # self is varibale holding (prem)
    print('GM')
    print(f'Hi : {self.r}')

prem = MECH()
prem.kumar()
```

```python
class CSE:
  a = 10
  b = 20

  def addition(self):
    print(self.a + self.b)

  def subtract(self):
    print(self.a - self.b)

  def common(self):
    self.addition()
    self.subtract()


anil = CSE()
anil.common()
```

```python
#Constructor

# Constructor is used for initalizing what we needed (Class name and Constructor name will be same)

# There are 2 types of Constructors

#       - `Non_Parameter Constructors`
#       - `Parameter Constructors`
```

```python
class MECH:
  f = 20

  def __init__(self):
    print('Good Morning')

anil = MECH()
```

```python
class MECH:
  f = 20

  def __init__(self,a):
    print('Good Morning',a)

anil = MECH(10)
```

```python
class CSE:
  def __init__(self,a,b):
    self.a = a
    self.b = b

  def addition(self):
    print(self.a + self.b)
    self.subtraction()

  def subtraction(self):
    print(self.a - self.b)

anil = CSE(1,2)
anil.addition()
```

```python
# without constructor

class CALCULATOR:
  a = 10
  b = 20
  def addition(self):
    print(self.a + self.b)

  def subtraction(self):
    print(self.a - self.b)

  def multiply(self):
    print(self.a * self.b)

  def division(self):
    print(self.a / self.b)

anil = CALCULATOR()
anil.addition()
anil.subtraction()
anil.multiply()
anil.division()
```

```python
# with constructor

class CALCULATOR:
  def __init__(self,a,b):
    self.a = a
    self.b = b

  def addition(self):
    print(self.a + self.b)

  def subtraction(self):
    print(self.a - self.b)

  def multiply(self):
    print(self.a * self.b)

  def division(self):
    print(self.a / self.b)

anil = CALCULATOR(100,200)
anil.addition()
anil.subtraction()
anil.multiply()
anil.division()
```

```python
import sys
```

```python

class PRIME:
  def __init__(self,a,b):
    try:
      self.a = a
      self.b = b
    except Exception as e:
      er_ty , er_ms , er_lin = sys.exc_info()
      print(f'Error in Line no : {er_lin.tb_lineno} : due to {er_ty}')

  def calculation(self):
    try:
      for i in range(self.a,self.b): # (1,101)
        c = 0
        for j in range(1,i+1):
          if i % j == 0:
            c = c + 1
        if c == 2:
          print(f'Prime : {i}')
        else:
          print(f'Not Prime : {i}')
    except Exception as e:
      er_ty , er_ms , er_lin = sys.exc_info()
      print(f'Error in Line no : {er_lin.tb_lineno} : due to {er_ty}')


anil = PRIME(1,10)
anil.calculation()
```

```python
class EVENODD:
  def __init__(self,a,b):
    try:
      self.a = a
      self.b = b
    except Exception as e:
      er_ty , er_ms , er_lin = sys.exc_info()
      print(f'Error in Line no : {er_lin.tb_lineno} : due to {er_ty}')
  def fun(self):
    try:
      for i in range(self.a,self.b):
        if i % 2 == 0:
          print(f'Even : {i}')
    except Exception as e:
      er_ty , er_ms , er_lin = sys.exc_info()
      print(f'Error in Line no : {er_lin.tb_lineno} : due to {er_ty}')

anil = EVENODD(11,16)
anil.fun()
```
