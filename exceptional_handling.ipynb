```python
#Since we are getting the Errors (which) is not in optimized format -> Using Exceptional Handling -> we can convert the Error into Optimized format
```

```python
try:
  a = 10 # save in the Memory
  b = 20 # save in the Memory
  print(b / a)


except Exception as e:
  print(e)
```

```python
try:
  a = 0 # save in the Memory
  b = 20 # save in the Memory
  print(b / a) # 8lines


except Exception as e:
  print(e)
```

```python
# finally is used as like print condition it can executed with or without error.
```

```python
try:
  a = 10
  b = 20
  print(a / b)

except Exception as e:
  print(e)


finally:
  print('I will run without Error')
```
