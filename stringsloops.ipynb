```python
# Write a program using a for loop to print each character of a string on a new line.

a = 'sunil'
for i in a:
  print(i)
```

```python
# Write a program using a while loop to print each character of a string with its index number.

a = 'sunil'
b = 0
while b < len(a):
    print(b, ":", a[b])
    b += 1
```

```python
# Write a for loop program to count how many vowels are in a given string.
a = 'sunil'
count = 0
for i in a:
  if i in 'aeiouAEIOU':
    count = count + 1
print(count)
```

```python
# Write a while loop program to reverse a string manually.
a =int(input())
b = 0
while a>0:
  c = a%10
  b=(b*10)+c
  a=a//10
print(b)
```

```python
# Write a for loop program that prints characters of a string except spaces.
a = 'sunil reddy'
for i in a:
  if i != ' ':
   print(i,end='')
```

```python
# Write a while loop program that stops printing characters when it encounters the letter a.
a = 'java'
s = 0
while s <len(a):
  if a[s]=='a':
    break
  print(a[s])
  s+=1
```

```python
# Write a program using a for loop to check how many digits exist inside a mixed string.
a = 'sunil12345reddy'
b = 0
for i in a:
  if i.isdigit():
    print(i)
    b=b+1
print(b)
```

```python
#Write a while loop program that prints the first half of a string.

a = 'sunilreddy'
b = int(len(a)/2)
while b>0:
  print(a[b])
  b=b-1
```

```python
fruits={'apple':100,'banana':80,'carrot':190}
k=list(fruits.keys())
v=list(fruits.values())
print(k)
print(v)
v.sort()
print(v)
p=[]

for i in v:
    for j in fruits:
        if fruits[j]==i:
            p.append(j)
print(p)

f = {}

for i in range(len(p)):
    f[v[i]]=p[i]

print(f)

```
