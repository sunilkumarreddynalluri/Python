```python
#using while loop
a = ['sunil','karthik']
i = 0
while i<len(a):
  print(a[i])
  i = i + 1
```

```python
#Write a for loop program to find the sum of all elements in a list.
a = [1,2,3,4,5]
b = 0
for i in a:
  b = b + i
print(b)
```

```python
#Write a while loop program to find the largest number in a list.
a = [1,2,3,4,5]
b = 0
large = a[0]
while b<len(a):
  a[b]>large
  large = a[b]
  b=b+1
print(large)
```

```python
#Write a for loop program to print only even numbers from a list.
a = [1,2,3,4,5,6,7,8]
b = []
for i in a:
  if i%2==0:
    b.append(i)
print(b)
```

```python
#Write a for loop program to count how many times a specific element appears in a list.

a = list(map(int, input("Enter numbers separated by spaces: ").split()))

elements = list(map(int, input("Enter elements to count: ").split()))

for e in elements:
    count = 0
    for i in a:
        if i == e:
            count = count + 1
    print("The element", e, "appears", count, "times.")

```

```python
#Write a while loop program to reverse a list without using the reverse function.
a=[12,14,15]
b=[]
c=len(a)-1
while c>=0:
  b.append(a[c])
  c=c-1
print(b)
```
